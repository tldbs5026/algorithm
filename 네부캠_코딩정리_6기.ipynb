2. K번째수

배열 array의 i번째 숫자부터 j번째 숫자까지 자르고 정렬했을 때, k번째에 있는 수를 구하려 합니다.

예를 들어 array가 [1, 5, 2, 6, 3, 7, 4], i = 2, j = 5, k = 3이라면

    array의 2번째부터 5번째까지 자르면 [5, 2, 6, 3]입니다.
    1에서 나온 배열을 정렬하면 [2, 3, 5, 6]입니다.
    2에서 나온 배열의 3번째 숫자는 5입니다.

배열 array, [i, j, k]를 원소로 가진 2차원 배열 commands가 매개변수로 주어질 때, commands의 모든 원소에 대해 앞서 설명한 연산을 적용했을 때 나온 결과를 배열에 담아 return 하도록 solution 함수를 작성해주세요.

```python
array = [1, 5, 2, 6, 3, 7, 4]
commands = [[2, 5, 3], [4, 4, 1], [1, 7, 3]]

def solution(array, commands) :
  answer = []

  for c in commands :
    i,j,k = c[0],c[1],c[2]
    arr = array[i-1:j]
    arr = sorted(arr)
    answer.append(arr[k-1])

  return answer

solution(array, commands)
```

5. 괄호 회전하기

```python
# s가 회전하면서 올바른 괄호인 경우를 찾는다
# '{[()}]'    # 올바른게 아닐듯

s = "}]()[{"
p = {')' : '(', ']' : '[', '}' : '{'}
def solution(s) :
  answer = 0

  for _ in range(len(s)) :
    inspect = []
    s = s[1:] + s[0]      # 왼쪽으로 회전
    ok = True

    for i in s :
      if i == '[' or i =='{' or i =='(' :   # 열려있을때
        inspect.append(i)
      else :      # 닫혀있을 때
        if len(inspect) == 0 :
          ok = False
          break

        if p[i] == inspect[-1] :
          inspect.pop()
        else :
          ok = False
          break

    if len(inspect) != 0 :
      ok = False
      break

    if ok == True :
      answer +=1
  return answer

solution(s)
```

6. n^2 배열 자르기

```python
def solution(n,left,right) :

  matrix = [[0 for _ in range(n)] for _ in range(n)]
  new_matrix = []

  for i in range(len(matrix)) :   # row
    for j in range(len(matrix[0])) :  # col
      new_matrix.append(max(i,j)+1)

  return new_matrix[left:right+1]
```

어떤 경우에 의해 방향을 회전하지 않고,
모든 경우에서 회전하는 것을 계산하여 목적지에 도착하기만 하면 ok.

```python
from collections import deque

board = [[0, 0, 0, 1, 1],
          [0, 0, 0, 1, 0],
            [0, 1, 0, 1, 1],
            [1, 1, 0, 0, 1],
              [0, 0, 0, 0, 0]]
n = len(board)
m = len(board[0])
# 상하좌우
dx = [0,0,-1,1]
dy = [-1,1,0,0]

q = deque()

# (x1,y1), (x2.y2)
position = ((0,0),(0,1))
pos = [position]
time = 0


q.append((pos,time))

while q :
  pos, time = q.popleft()

  if ((n-1,m-1)) in position :
    return time

  for candi in next_pos(board,pos) :
    if candi in pos :
      continue

    pos.append(candi)
    q.append((candi, time+1))

def next_pos(board,pos) :
  lst = list()

  ((x1,y1),(x2,y2)) = pos

  for dir in range(4) :
    nx1,ny1, nx2,ny2 = x1 + dx[dir] , y1+ dy[dir], x2 + dx[dir], y2 + dy[dir]

    #밖으로 나가면 continuje
    if nx1 < 0 or nx1 >= n or ny1 <0 or ny1 >= m :
      continue
    if nx2 < 0 or nx2 >= n or ny2 <0 or ny2 >= m :
      continue
    # board의 값이 0이 아니면 continue
    if board[nx1][ny1] or board[nx2][ny2] :
      continue

    lst.append(set((nx1,ny1),(nx2,ny2)))

  for rot in range(2) :
    if y1 == y2 :
      ny1,ny2  = y1+dy[rot], y2+dy[rot]
      if not (ny1 < 0 or ny1 >= m or ny2 < 0 or ny2 >= m) :
        if not (board[x1][ny1] or board[x2][ny2]) :
          lst.append(set((x1,y1),(x2,ny2)))
          lst.append(set((x1,ny1),(x2,y2)))

    elif x1 == x2 :
      nx1,nx2 = x1+dx[rot+2], x2+dx[rot+2]
      if not (nx1 < 0 or nx1 >= n or nx2 <0 or nx2 >= n ) :
        if not (board[nx1][y1] or board[nx2][y2]) :
          lst.append(set((nx1,y1),(x2,y2)))
          lst.append(set((x1,y1),(nx2,y2)))

  return lst
```

```python
# queue를 써도 되고 최솟값 기본정렬이 되는 heap을 써도되고

import heapq

def solution(operations) :
  heap = []

  for ops in operations :
    comm, number = ops.split(' ')

    if comm == 'I' :
      heapq.heappush(heap, int(number))
    # 가장 큰 값을 갖는 heap안에서의 index에 해당하는 값을 pop
    elif comm == 'D' and heap :
      if number =='1' :
        heapq.pop(heap.index(heapq.nlargest(1, heapq[0])))
      else :
        # 최솟값 heappop
        heapq.heappop(heap)

  # 모든 연산이 끝나고
  if not heap :
    return [0,0]

  return [heap.index(heapq.nlargest(1, heapq[0])), heappop(heap)]


```

```python
dp = [0] * (n+1)


def solution(n,money) :
  # 각 단위 동전마다 가지수 추가
  for coin in money :
    dp[coin] +=1
    # 그 가지수 부터 시작하여 만들수 있는 dp테이블
    for price in range(coin, n+1) :
      dp[price] += dp[price-coin]

  return dp[n]
```
