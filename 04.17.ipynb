```python
from itertools import permutations as perm
from collections import deque

expression = "100-200*300-500+20"

def solution(expression):
  answer = 0
  for operation in perm(['+','-','*'] ,3) :
    answer = max(answer, abs(equation(operation,expression)))
  return answer

def equation(operation, expression ) :
  q = deque()
  exp = ''

  for e in expression :
    if e.isdigit() :
      exp += e
    else :
        q.append(exp)
        exp = ''
        q.append(e)

    for op in operation :
      
      while len(q) != 0 :
        s = q.popleft()
        cal = s[0]
        if s == op :
          result = eval(cal.pop() + op + q.popleft())
          cal.append(result)
        else :
          cal.append(s)
      q = deque(cal)
  return int(q.pop())

solution(expression)
```

```python
from itertools import combinations

def solution(orders, course) :
  answer = []

  for c in course :
    candidates = []
    new_menu = {}

    for menu in orders :
      menu_list = list(''.join(menu))
      for lst in combinations(menu_list, c) :
        res = list(''.join(sorted(lst)))
        if res not in candidates :
          candidates.append(res)

        else :
          if res not in new_menu.keys() :
            new_menu[res] = 2
          else :
            new_menu[res] +=1
        
      
      sort_menu = sorted(new_menu.items(), key=lambda x  : [len(x[0],x[1])])

      if len(sort_menu) :
        now = course[-1]
        maximum = sort_menu[-1][1]

      while sort_menu :
        menu, cnt = sort_menu.pop()
        if len(menu) == now and cnt >= maximum :
          answer.append(menu)
        
        elif len(menu) != now :
          now = len(menu)
          maximum = cnt
          answer.append(menu)
    
    return sorted(answer)
```

```python
from itertools import combinations
from collections import Counter

def solution(orders, course) :
  answer = []

  for k in course :
    candidates = []

    for menu_lst in orders :
      for lst in combinations(menu_lst, k) :
        res = ''.join(sorted(lst))
        candidates.append(res)
      
      sorted_candidates = Counter(candidates).most_common()
      answer += [menu, cnt in sorted_candidates if cnt > 1 and cnt == sorted_candidates[0][1]]
    
  return sorted(answer)
```

```python
places = [["POOOP", 
           "OXXOX", 
           "OPXPX", 
           "OOXOX", 
           "POXXP"], 
          ["POOPX", "OXPXP", "PXXXO", "OXXXO", "OOOPP"], 
          ["PXOPX", "OXOXP", "OXPOX", "OXXOP", "PXPOX"], 
          ["OOOXX", "XOOOX", "OOOXX", "OXOOX", "OOOOO"], 
          ["PXPXP", "XPXPX", "PXPXP", "XPXPX", "PXPXP"]]

from collections import deque

def bfs(p) :
  start = []

  for i in range(5) :
    for j in range(5) :
      if p[i][j] == 'P' :
        start.append([i,j])

  for s in start :
    queue = deque([s])
    visited = [[False] * 5 for _ in range(5)]
    distance= [[0] * 5 for _ in range(5)]
    visited[s[0]][s[1]] = True

    while queue :
      x,y = queue.popleft()

      dx = [0,0,-1,1]
      dy = [-1,1,0,0]

      for i in range(len(dx)) :
        nx = x + dx[i]
        ny = y + dy[i]

        if 0<=nx<=5 or 0<=ny<=5 or visited[nx][ny] == False :

          if p[nx][ny] == 'O' :
            queue.append([nx,ny])
            visited[nx][ny] = True
            distance[nx][ny] = distance[x][y] + 1

          if p[nx][ny] =='P' and distance[nx][ny] <= 1 :
            return False
    return True

def solution(places) :
  answer =[]

  for i in places :
    answer.append(bfs(i))
  return answer
```

```python
from collections import defaultdict
from bisect import bisect_left

def solution(info, query) :
  dic = defaultdict(
      lambda x : defaultdict(
          lambda x : defaultdict(
              lambda x : defaultdict(list)
          ),
      ),
  )
  for item in info :
    item = item.split()
    dic[item[0]][item[1]][item[2]][item[3]].append(int(item[4]))
  
  def score_check(d,q) :
    result = 0
    d.sort()
    return len(d) - bisect_left(d, int(q[4]))
  def pc_check(d,q) :
    if q[3] == '-' :
      result = 0
      for k in d.keys() :
        result += score_check(d[k],q)
    return result
  
  def js_check(d,q) :
    if q[2] == '-' :
      result = 0
      for k in d.keys() :
        result += pc_check(d[k],q)
      return result
    else :
      return pc_check(d[q[1]],q)
  
  def language(d,q) :
    if q[0] == '-' :
      result = 0
      for k in d.keys() :
        result += js_check(d[k],q)
      return result
    else :
      return js_check(d[q[0]],q)
  result = []


  for st in query :
    q = st.split(' and ')
    tmp = q[-1].split(' ')
    q[-1] = tmp[0]
    q.append(tmp[1])
    result.append(language(dic,q))

  return result
```

```python
from collections import defaultdict
from itertools import product
from bisect import bisect_left

def solution(info, query) :
  dic = defaultdict(list)

  for item in info :
    item = item.split()
    slot = []

    for i in range(4) :
      slot.append({'-', item[i]})

    for i in product(*slot) :
      dic[''.join(i)].append(int(item[4]))
    
  for k in dic.keys() :
    dic[k].sort()
  
  result = []
  for q in query :
    q = q.replace(' and ', ' ').split(' ')
    score = int(q[-1])
    q = q[:-1]

    t = dic[''.join(q)]
    result.append(len(t) - bisect_left(t, score))
  
  return result
```
