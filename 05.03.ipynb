```python
from collections import defaultdict
want = ["banana", "apple", "rice", "pork", "pot"]
number = [3, 2, 2, 2, 1]
discount = ["chicken", "apple", "apple", "banana", "rice", "apple", "pork", "banana", "pork", "rice", "pot", "banana", "apple", "banana"]


def solution(want, number, discount):
    cnt =0
    want_dic = {}

    #want랑 number를 합친 dict
    for i in range(len(want)) :
      want_dic[want[i]] = number[i]
    want_dic = sorted(want_dic.items(), key = lambda x : x[0])
    
    # print('1', want_dic)
    #discount를 돌며
    for i in range(len(discount)) :
        lst = defaultdict(int)
        

        #discount안에 10개
        dis_lst = discount[i:i+10]
        if len(dis_lst) != 10 :
          break

        # print(dis_lst)
        # j가 want에 있으면 카운트 추가
        for j in dis_lst :
            if j in want :
              lst[j] +=1
        

        lst = sorted(lst.items(), key = lambda x : x[0])

        if want_dic == lst :
          cnt +=1

    return cnt

solution(want, number, discount)
```

```python
#bfs

def solution(numbers, target) :
  answer = 0
  leaves = [0]

  for num in numbers :
    tmp = []

    for parent in leaves :
      tmp.append(parent + num)
      tmp.append(parent - num)
    
    leaves = tmp
  
  for leaf in leaves :
    if leaf == target :
      answer +=1

  return answer
```

```python
# dfs

def solution(numbers, target) :
  answer = dfs(numbers, target,0)
  return answer


def dfs(numbers,target, depth) :
  answer = 0
  if depth == len(numbers) :
    print(numbers)
    if sum(numbers) == target :
      return 1
    
    return 0
  
  else :
    # +일때 한번
    answer += dfs(numbers,target, depth +1)
    # -일때 한번
    numbers[depth] *= -1
    answer += dfs(numbers,target, depth+1)
    return answer
```
