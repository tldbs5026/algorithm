###     최대공약수와 최소공배수



```python
def gcd(n,m,a) :
    if n == 0 :
        return m
    if m == 0 :
        return n

    for i in range(1, min(n,m)+1) :
        if n%i ==0 and m%i == 0 :
            n //=i
            m //=i
            a *=i

    return a

def lcm(n,m,b) :
    if n == 0 and m == 0 :
        return 0
    for i in range(1, max(n,m)+1) :
        if n%i ==0 and m%i == 0 :
            n //=i
            m //=i
            b *=i

    return b*n*m

def solution(n, m):
    a=1
    b=1

    answer = [gcd(n,m,a), lcm(n,m,b)]
    return answer
```

###  3진법 뒤집기

https://school.programmers.co.kr/learn/courses/30/lessons/68935

n,b = divmod(n,3) = n을 3으로 나누어서 나온 몫(n)과 나머지(b) 반환

진법은 기본적으로 나머지를 기준으로 전개한 것이다.

또한 int는 단순히 숫자형으로 바꾸는 것이 아니라 뒤에 조건을 붙임으로써 3 -> 10진법으로 전환이 가능함

```python
def solution(n):
    answer = 0
    third = ''

    while n >= 1 :
        n,b = divmod(n,3)
        third += str(b)

    answer = int(third, 3)
    return answer



```

```python
s = "AB"
n =1

def solution(s, n):
    s=list(s)
    for i in range(len(s)) :
        if s[i].islower() :
            s[i] = chr( (ord(s[i]) - ord('a') + n) % 26 + ord('a') )

        elif s[i].isupper() :
          s[i] = chr( (ord(s[i]) - ord('A') + n) % 26 + ord('A') )

    return ''.join(s)

solution(s,n)
```

```python
sizes = [[60, 50], [30, 70], [60, 30], [80, 40]]
def solution(sizes):
    answer = 0
    w_max = 0
    h_max = 0
    for w,h in sizes :

        if w > h :
            w,h = h,w
        w_max = max(w_max, w)
        h_max = max(h_max, h)
    answer = w_max * h_max
    return answer

solution(sizes)
```

```python
food = [1, 3, 4, 6]


def solution(food):
    lst = []
    answer = ''

    for i in range(1, len(food)) :
        if food[i] > 1 and food[i] %2 == 0 :
            lst.append( str(i) * (food[i]//2))
        else :
            lst.append( str(i) * ((food[i]-1)//2))

    lst.append('0')
    rev = reversed(lst[:-1])

    answer = ''.join(lst)
    answer += ''.join(rev)

    return answer

solution(food)
```
