```python
players = ["mumu", "soe", "poe", "kai", "mine"] 	
callings = ["kai", "kai", "mine", "mine"] 	

player_dic = {player : rank for rank, player in enumerate(players)}

# {'mumu' : 0, 'soe' : 1 ...}

rank_dic = {rank : player for rank, player in enumerate(players)}





def solution(players, callings):
  for call in callings :
    now = player_dic[call]
    player_dic[rank_dic[now-1]], player_dic[rank_dic[now]] = player_dic[rank_dic[now]], player_dic[rank_dic[now-1]]
    rank_dic[now-1], rank_dic[now] = rank_dic[now], rank_dic[now-1]

  return list(rank_dic.values())

solution(players, callings)
```

```python
from collections import defaultdict
today = "2022.05.19"
terms = ["A 6", "B 12", "C 3"]
privacies = ["2021.05.02 A", "2021.07.01 B", "2022.02.19 C", "2022.02.20 C"]



def solution(today, terms, privacies):
    today = calcu(today)
    dic = defaultdict()

    for term in terms :
      alp, m = term.split()
      dic[alp] = int(m) * 28
    
    answer = []

    for i in range(len(privacies)) :
      collected_day, alp = privacies[i].split()
      if today >= dic[alp] + calcu(collected_day) :
        answer.append(i+1)
    return answer


def calcu(today) :
  y,m,d = map(int, today.split('.'))
  return (y-2000) * 28 * 12 + m * 28 + d

```

```python
def solution(cap, n, deliveries, pickups):
    deliveries = deliveries[::-1]
    pickups = pickups[::-1]
    answer = 0

    d = 0
    p = 0

    for i in range(n) :
      d += deliveries[i]
      p += pickups[i]

      while d > 0 and p > 0 :
        d -= cap
        p -= cap
        answer += (n-i) * 2
    return answer
  
```

```python
from itertools import permutations
babbling = ["aya", "yee", "u", "maa", "wyeoo"]
lst =  ["aya", "ye", "woo", "ma"]
def solution(babbling) :
  answer = 0
  

  word_lst = []
  for i in range(1, len(lst) +1) :
    for j in permutations(lst,i) :
      word_lst.append(''.join(j))


  for i in babbling :
    if i in word_lst :
      answer +=1

  return answer

   
```

```python
# 1단계 new_id의 모든 대문자를 대응되는 소문자로 치환합니다.
# 2단계 new_id에서 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거합니다.
# 3단계 new_id에서 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환합니다.
# 4단계 new_id에서 마침표(.)가 처음이나 끝에 위치한다면 제거합니다.
# 5단계 new_id가 빈 문자열이라면, new_id에 "a"를 대입합니다.
# 6단계 new_id의 길이가 16자 이상이면, new_id의 첫 15개의 문자를 제외한 나머지 문자들을 모두 제거합니다.
#      만약 제거 후 마침표(.)가 new_id의 끝에 위치한다면 끝에 위치한 마침표(.) 문자를 제거합니다.
# 7단계 new_id의 길이가 2자 이하라면, new_id의 마지막 문자를 new_id의 길이가 3이 될 때까지 반복해서 끝에 붙입니다.
import re

new_id = "123_.def"

def solution(new_id):
  new_id = new_id.lower()
  new_id = re.sub(r'[^a-z0-9\-\_\.]', '', new_id)
  new_id = re.sub(r'[\.]+','.', new_id)
  if new_id.startswith('.') :
    new_id = new_id[1:]
  if new_id.endswith('.') :
    new_id = new_id[:-1]
  if len(new_id) == 0 :
    new_id ='aaa'
  if len(new_id) >= 16 :
    new_id = new_id[:15]
    if new_id.endswith('.') :
      new_id = new_id[:-1]
  if len(new_id) <= 2 :
    while len(new_id) < 3 :
      new_id +=new_id[-1]
  return new_id

  
solution(new_id)
```

```python
people = [70, 50, 80, 50]
limit = 100

r = len(people)
lst=[False] * len(people)
cnt=0

for i in range(r) :
  if lst[i] != True :
    lst[i] = True
    lim = limit - people[i]
    for j in range(i+1,r) :
      if lst[j] != True :
        if lim - people[j] >= 0 :
          lst[j] = True
          lim -= people[j]
        if lim <= 0 :
          cnt +=1
          lim = limit
          break
      
      
```
