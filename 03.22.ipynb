```python
#문제 : 두 개로 나누어진 네트워크에 대한 송전탑의 개수 최소화
## 즉, bfs(혹은  dfs로 개수를 구한 후 , 최솟값에 대한  dp)

from collections import deque

def solution(n,wires) :
  #wires를 graph로 재정의
  graph = [[] for _ in range(n+1)]

  for v1,v2 in wires :
    graph[v1].append(v2)
    graph[v2].append(v1)



  #bfs를 정의
  def bfs(start) :
    q = deque([start])

    visited = [False] * (n+1)
    visited[start] = True

    cnt = 1

    while q :
      s = q.popleft()

      for i in graph[s] :
        if not visited[i] :
          q.append(i)
          visited[i] = True
          cnt +=1

    return cnt


  #조립
  ## 먼저 각 네트워크를 분리
  answer = n

  for v1,v2 in wires :
    graph[v1].remove(v2)
    graph[v2].remove(v1)

    answer = min(answer, abs(bfs(v1) - bfs(v2)) )

    #분리시켰던 네트워크 재조립

    graph[v1].append(v2)
    graph[v2].append(v1)


  return answer




#이렇게 주어졌을 때,
n = 7
wires = [[1,2],[2,7],[3,7],[3,4],[4,5],[6,7]]

solution(n,wires)




```

```python
##줄 서는 방법의 총 개수
  # factorial을 사용하기 때문에 이것에 대한 정의 필요

def factorial(n) :
  if n <2 :
    return 1

  return n * factorial(n-1)

  #각 자리수에 해당하는 수의 개수는 각각 factorial(n-1)


def solution(n,k) :
  answer = []
  graph = [i for i in range(1,n+1)]

  #idx를 지정(n=3일때, [1,2,3]이 저장된)


  while n != 0 :
    rank = factorial(n-1)
    idx = k // rank
    #다음 자리수에서는 k의 개수가 감소되어야함
    k = k % rank

    if k != 0 :
      answer.append(graph.pop(idx))
    else :
      answer.append(graph.pop(idx-1))

    n-=1


  return answer


n = 3
k = 5
solution(n,k)


```

```python
# genres[i]는 고유번호가 i인 노래의 장르입니다.
# plays[i]는 고유번호가 i인 노래가 재생된 횟수입니다.

# 노래를 수록하는 기준
# 1. 장르 (genre)
# 2. 장르 내에서 재생 수 (plays)
# 3. 장르 내의 재생 수에서 낮은 고유번호 (i)


def solution(genres, plays) :
  total = []
  answer = []
  #total에 장르와 재생 수, 고유번호를 정리
  total = [[genres[i], plays[i], i] for i in range(len(genres))]
  total = sorted(total, key = lambda x  : (x[0], -x[1], x[2]))

  #각각 중복되는 장르를 통합하고, 재생 수를 더하여 return할 순서를 정하기

  total_genre = {}

  for i in total :
    if i[0] not in total_genre :
      total_genre[i[0]] = i[1]
    else :
      total_genre[i[0]] += i[1]

  total_genre = sorted(total_genre.items(), key=lambda x : -x[1])

  for i in total_genre :
    cnt =0

    for j in total :
      if i[0] == j[0] :
        cnt +=1

        if cnt >2 :
          break

        else :
          answer.append(j[2])

  return answer

genres = ["classic", "pop", "classic", "classic", "pop"]
plays = [500, 600, 150, 800, 2500]

solution(genres, plays)
```

```python
def solution(n,money) :
  dp =  [0] *(n+1)

  for coin in money :
    dp[coin] +=1
    for j in range(coin, n+1) :
      dp[j] += dp[j-coin]

  answer = dp

  return answer[n]

n=5
money = [1,2,5]

solution(n,money)


```
